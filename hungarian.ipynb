Grafik diatas persebaran jumlah target tidak seimbang oleh karena itu perlu diseimbangkan terlebih dahulu dengan 2 cara yaitu `Oversampling` dan `Undersampling`.

Oversampling merupakan teknik dari pengolahan data untuk mengatasi ketidakseimbangan kelas dalam dataset.
Oversampling dilakukan jika jumlah dataset sedikit sedangkan undersampling dilakukan jika jumlah data terlalu banyak.
Disini kita akan melakukan oversampling dikarenakan jumlah data kita tidak banyak.
Salah satu metode yang Oversampling yang akan kita gunakan adalah SMOTE


Mengimport SMOTE

### 7) Modeling

Pada tahap ini kita akan memulai untuk membangun sebuah model.

Dibawah ini merupakan sebuah fungsi untuk menampilkan hasil akurasi dan rata - rata dari recall , f1 dan precision score setiap model. Fungsi ini nantinya akan dipanggil di setiap model. Membuat Fungsi ini bersifat opsional.


```python
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score,roc_auc_score,confusion_matrix,precision_score
```

```python
def evaluation(Y_test,Y_pred):
    acc = accuracy_score(Y_test,Y_pred)
    rcl = recall_score(Y_test,Y_pred,average = 'weighted')
    f1 = f1_score(Y_test,Y_pred,average = 'weighted')
    ps = precision_score(Y_test,Y_pred,average = 'weighted')

    metric_dict={'accuracy': round(acc,3), 'recall': round(rcl,3),
                    'F1 score': round(f1,3),
                    'Precision score': round(ps,3)
                    }

    return print(metric_dict)
```

Yang merupakan langkah awal untuk memahami bagaimana setiap algoritma berperilaku dalam menangani data dan membuat prediksi.Langkah awal mencakup penanganan ketidakseimbangan data dengan menggunakan teknik random over sampling SMOTE guna meningkatkan keseimbangan antara kelas mayoritas dan minoritas dalam dataset. Selain itu, proses normalisasi data dilakukan untuk memastikan konsistensi skala pada fitur-fitur yang digunakan. Dilakukan juga proses tuning parameter pada masing-masing algoritma menggunakan RandomSearchCV untuk mencari kombinasi parameter yang optimal agar model dapat lebih baik dalam melakukan prediksi. Hasil dari pengujian menunjukkan bahwa klasifikasi menggunakan Random Over Sa,pling SMOTE pada model KNN menghasilkan akurasi sebesar 75.4%, model Random Forest dengan akurasi 92%, dan model XGBoots mencapai 90.4%. Kemudian, saat menggunakan data yang sudah dinormalisasi dan Random Over Sampling SMOTE pada model KNN, terjadi pennigkatan akurasi menjadi 86.1%, sedangkan pada model Random Forest dan XGBoots tidak ada perubahan signifikan dalam akurasi. Namun, setelah dilakukan tuning menggunakan RandomSearchCV, normalisasi, dan Random Over Sampling SMOTE, terjadi peningkatan signifikan pada model KNN mencapai akurasi 93%, sedangkan pada model Random Forest mengalami penurunan akurasi menjadi 87.7%, dan model XGBoots tetap konsisten dengan akurasi 92%. Secara keseluruhan, penanganan ketidaksiembangan data dengan menggunakan kombinasi metode oversampling, tuning parameter, dan normalisasi menunjukkan dampak yang signifikan terhadap performa model klasifikasi, khususnya pada model KNN dan XGBoots. Ini menekankan pentingnya pemilihan model terbaik dan parameter yang optimal untuk meningkatkan akurasi dan kinerja model secara keseluruhan dalam menangani ketidakseimbangan data.
